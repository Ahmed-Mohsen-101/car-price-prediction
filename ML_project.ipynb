Training model

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=df.drop(columns=['selling_price'])
y=df['selling_price']
```

```python
ordinal_encoder=OrdinalEncoder()
x['owner']=ordinal_encoder.fit_transform(x[['owner']])
```

```python
x_encoded=pd.get_dummies(x,columns=['name','fuel','seller_type','transmission'],drop_first=True)
```

```python
# Select boolean columns (usually dummies)
bool_cols = x_encoded.select_dtypes(include='bool').columns

# Convert only those to int
x_encoded[bool_cols] = x_encoded[bool_cols].astype(int)
x_encoded
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_encoded,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
num_cols = ['year', 'km_driven', 'seats', 'max_power (in bph)', 'Mileage', 'Engine (CC)']
x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
x_train
# scaler=StandardScaler()
# x=scaler.fit_transform(x_encoded)
# x
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
rmse
```

```python
import matplotlib.pyplot as plt

# y_test = actual car prices
# y_pred = predicted prices from the model

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.6, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', linewidth=2)  # perfect prediction line
plt.xlabel('Actual Selling Price')
plt.ylabel('Predicted Selling Price')
plt.title('Actual vs Predicted Car Prices')
plt.show()
```

```python
from sklearn.metrics import r2_score

# Calculate R2
score = r2_score(y_test, y_pred)
print(f"Model Accuracy (R2 Score): {score * 100:.2f}%")
```
